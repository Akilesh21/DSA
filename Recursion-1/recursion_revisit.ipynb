<h2>Power of number</h2>

```python
def power_n(n,m):
    # n = number of which we want the power
    # m = raised the n
    return n**m
n,m = input().split()
n = float(n)
m = int(m)
power_n(n,m)    
power_n(n,m)    
```

check number in array


```python
def check_number(arr,x):
    if len(arr)==0:
        return False
    small_arr = arr[1:]
    if x == arr[0]:
        return True
    else:
        return check_number(small_arr,x)
arr = [int(i) for i in input().split()] 
x = int(input())
if check_number(arr,x):
    print("true")
else:
    print("false")       
```

last index of array

```python
def lastIndex(a,x):
    l=len(a)
    if l==0:
        return -1
    smallerList=a[1:]
    smallerListOutput=lastIndex(smallerList,x)
    if smallerListOutput!=-1:
        return smallerListOutput+1
    else:
        if a[0]==x:
            return 0
        else:
            return -1

    
arr = [9,8,10,8]
print(lastIndex(arr,8))



    
```

remove X

```python
def removeX(string,ans): 
    if len(string) == 0:
        return ''
    if string[0] == 'x':
        return ''
    ans =string[0] + removeX(string[1:],ans)
    return ans
    

# Main
string = input()
new_S = ''
print(removeX(string,new_S))
```
