<h1>Inheritance</h1>

```python
#parent class
class vehicle():
    def __init__(self,name,model,fuel):
        #parent attribute
        self.name = name
        self.model = model
        self.fuel = fuel
# child class        
class car(vehicle):
    def __init__(self, name, model, fuel,air_conditioning,sunroof):
        #parent attribute use this
        vehicle.name = name  
        vehicle.model = model
        vehicle.fuel = fuel
        #child attribute
        self.air_conditioning = air_conditioning
        self.sunroof = sunroof
    def show_parent_attribute(self):
        print(vehicle.name ," ",vehicle.model," ",vehicle.fuel)        
```

```python
obj = car("Tesla",2021,"electric","True","True")
obj.__dict__    
obj.show_parent_attribute()
```

<h1>Inheritace and Private members</h1>

```python
#parent class
class vehicle2():
    def __init__(self,name,model,fuel):
        #parent attribute
        self.name = name
        #private attributes
        self.__model = model
        self.__fuel = fuel
    #private function    
    def __private_method(self):
        print("This is private")
# child class        
class car2(vehicle2):
    def __init__(self, name, model, fuel,air_conditioning,sunroof):
        #parent attribute use this
        vehicle2.name = name  
        vehicle2.__model = model
        vehicle2.__fuel = fuel
        #child attribute
        self.air_conditioning = air_conditioning
        self.sunroof = sunroof
    def show_parent_attribute(self):
        print(vehicle2.name ," ",vehicle2.model," ",vehicle2.fuel)    
    def show_private_method_in_vehicle(self):
        self._vehicle2__private_method()  
obj2 = car2("bmw",2024,"diesel","True","True")
```

```python
obj2.show_private_method_in_vehicle()
```

<h2>Inheritance Continue</h2>

```python
class Exam:
    def __init__(self,subjects_marks,outoff):
        self.subjects_marks = subjects_marks
        self.outoff = outoff
    def exam(self):
        result = self.subjects_marks/self.outoff
        return result
class student(Exam):
    def __init__(self, subjects_marks, outoff,name):
        super().__init__(subjects_marks, outoff)
        self.name  = name
    def percentage(self):
        percentage = Exam.exam(self)*100
        return f"{self.name} your percentage is {percentage}"
```

```python
student1 =student(635,800,"Akhilesh soni") 
student1.percentage()
```

<h1>Polymorphism</h1>

```python
class human:
    def __init__(self,name):
        self.name = name
    def info(self):
        return f"hello my name is {self.name}"
class Male(human):
    def __init__(self, name):
        super().__init__(name)
    def info(self):
        return f"{human.info(self)} and i am a Male"
class Female(human):
    def __init__(self, name):
        super().__init__(name)
    def info(self):
        return f"{human.info(self)} And i am a female" 

```

```python
male = Male("Akhilesh")
male.info()
```

```python
female = Female("Jaya")
female.info()
```

<h2>Modifiers</h2>

```python
class Exam:
    def __init__(self,name,subjects,outoff):
        # public modifiers require no complexity to access we can access easily
        self.subjects = subjects
        self.__outoff = outoff
    def result(self):
        final_result = (self.subjects/self.__outoff)*100
        return final_result     
class Teacher(Exam):
    def __init__(self, name, subjects, outoff,add_marks):
        super().__init__(name, subjects, outoff)
        #private modifers have '__' it is hard to access require specific class name and same keyword to access
        # self.__add_marks = add_marks
        #protected modifiers have '_' not hard then private to access only require same keyword to access
        self._add_marks = add_marks
    def re_result(self):
        final_result = ((self.subjects+self._add_marks)/self.__outoff)*100
        return final_result
```

```python
student1 =Teacher("Akhilesh soni",635,800,1)
print(student1.__dict__)
student1.re_result()        
student1.result()
```

<h1>Multilevel Inheritance</h1>

```python
class A:
    def __init__(self,name):
        self.name = name
class B(A):
    def __init__(self, name,age):
        super().__init__(name)   
        self.age = age
class C(B):
    def __init__(self, name, age):
        super().__init__(name, age)
    def print(self):
        return self.age
n = C("akhilesh",21) 
n.print()      
```

<h1>Multiple and Method resolution order</h1>

```python
# Multiple inheritance
class Car():
    def __init__(self,make,model,fuel):
        self.make=make
        self.model=model
        self.fuel=fuel
    def get_car_detail(self):
        return "the make of the car is", self.make,"from car class"
    
class ElectricCar():
    def __init__(self,make,model,fuel):
        self.make=make
        self.model=model
        self.fuel=fuel
        
    def get_car_detail(self):
        return "the make of the car is",self.make,"from Electric Car class"

class Taxi(ElectricCar,Car):#Multiple Inheritance
    def __init__(self,make,model,fuel):
        super().__init__(make,model,fuel)
```

```python
myobj=Taxi("Tesla",2019,"electric")

myobj.get_car_detail()
```

```python
# Method resolution order (mro) is used what is the order of inheritance to understand what is the order of the method in the specific class
Taxi.__mro__
```

```python
Taxi.mro()
```

<h1>Operator overloading</h1>

```python
class Vehicle:
    def __init__(self,make,model,fare):
        self.make=make
        self.model=model
        self.fare=fare
    def __str__(self):
        return "The make of the model is {} and the model is of {} with a fare of {}".format(self.make,self.model,self.fare)
    def __add__(self,other):
        return self.fare+other.fare
```
