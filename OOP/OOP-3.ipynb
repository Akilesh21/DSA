<h1>abstract classes</h1>

```python
from abc import ABC,abstractmethod
class socialmedia(ABC):
    @abstractmethod
    def create_account(self):
        pass
    @abstractmethod
    def add_friends(Self):
        pass
class contribute(socialmedia):
    def create_account(self,name,email,age,phone_no):
        self.name = name 
        self.email = email
        self.age = age
        self.phone_no = phone_no
        return super().create_account()    
    def add_friends(self,friend_name):
        self.friend_name = friend_name 
        return super().add_friends()    
```

```python
user = contribute()
user.create_account("akhilesh soni","akhilesh@meta.com",21,46871313571)
user.age
```

<h1>one more example on abstract classes</h1>

```python
from abc import ABC,abstractmethod
class TextReaderAbstract(ABC):
    def __init__(self,path,filename):
        self.path=path
        self.filename=filename
    @abstractmethod #this is just an interface or "RULE"
    def get_path(self):
        pass
    @abstractmethod
    def get_filename(self):
        pass


class TextReader(TextReaderAbstract):
    def get_path(self):
        return self.path
    
    def get_filename(self):
        return self.filename
```

```python
myobj=TextReader("/user/download","sample.txt")
myobj.get_path()
```

```python
myobj.get_filename()
```
