```python
class practice:
    company = "google"
    def __init__(self,name,age,dob,number,skills):
        self.name = name
        self.age = age
        self.dob = dob
        #private
        self.__number = number
        self.__skills = skills
    def first(self):
        print(self.name)                
    @staticmethod
    def second():
        name = "akhilesh"
        return name
    @classmethod
    def third(self):
        age = 45
        return age    
```

```python
class calculator:
    def sum(self,n1,n2):
        return n1+ n2
    def sub(self,n1,n2):
        return n1- n2 
    def mul(self,n1,n2):
        return n1*n2
    def div(self,n1,n2):
        return n1/n2 
    def modul(self,n1,n2) :
        return n1%n2                 
```

```python
n = calculator()
n.sum(4,5)
```

```python
n.sub(11,8)
```

```python
n.mul(6,8)
```

```python
round(n.div(9,7),5)
```

```python
n.modul(10,4)
```
