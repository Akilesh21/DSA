```python
class college:
    def __init__(self,name,age,dept):
        super(college,self).__init__()
        self.name = name
        self.age = age
        self.dept = dept
class student(college):
    def __init__(self, name, age, dept,roll_no):
        super(student,self).__init__(name, age, dept)
        self.roll_no = roll_no
    def info(self):
        print(self.name," ",self.age," ",self.dept," ",self.dept)
s = student("Akhilesh soni",21,"comps",60)
s.info()                                
```
