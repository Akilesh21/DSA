```python
# Stack using array
class Stack:
    def __init__(self) -> None:
        self.__array = []
    def push(self,ele):
        self.__array.append(ele)
    def pop(self):
        if self.isEmpty():
            print("Stack is already empty can't pop")
            return   
        return self.__array.pop()        
    def size(self):
        return len(self.__array)
    def isEmpty(self):
        return self.size() == 0
    def top(self):
        if self.isEmpty():
            print("Stack is Empty")
            return
        return self.__array[self.size()-1]
    def show(self):
        return self.__array


```
