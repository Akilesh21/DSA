<h2>Queue</h2>

```python
class Queue:
    def __init__(self):
        self.__arr = []
        self.__count =0
        self.__front =0
    def enqueue(self, ele):
        self.__arr.append(ele)
        self.__count +=1
    def dequeue(self):
        if self.isEmpty():
            return -1
        ele = self.__arr[self.__front]
        self.__front +=1
        self.__count -=1
        return ele         
    def front(self):
        if self.__count==0:
            return -1
        return self.__arr[self.__front]
    def size(self):
        return self.__count
    def isEmpty(self):
        return self.size()==0
    def q(self):
        return self.__arr
    
q=Queue()
q.enqueue(1)
q.enqueue(2)
q.enqueue(3)
q.enqueue(4)
    
```

Implement Queue using Two stacks

```python
class MyQueue:

    def __init__(self):
        self.__s1 = []
        self.__s2 = []
        self.count = 0

    def push(self, x: int) -> None:
        while(len(self.__s1)!=0):
            self.__s2.append(self.__s1.pop())
        self.__s1.append(x)
        
        while(len(self.__s2)!=0):
            self.__s1.append(self.__s2.pop())
        return

    def pop(self) -> int:
        if len(self.__s1)==0:
            return -1
        return self.__s1.pop()  
    def peek(self) -> int:
        if len(self.__s1)==0:
            return -1
        return self.__s1[-1]

    def empty(self) -> bool:
        return len(self.__s1) == 0


# Your MyQueue object will be instantiated and called as such:
obj = MyQueue()
obj.push(1)
param_2 = obj.pop()
param_3 = obj.peek()
param_4 = obj.empty()
print(param_2)
print(param_3)
```

<h2>Reverse the queue</h2>

```python
def Reverse(queue):
    r_queue = []
    while queue.isEmpty() is False:
        ele = queue.dequeue()
        r_queue.append(ele)
    for i in range(0,len(r_queue)):
        queue.enqueue(r_queue[-i-1])
    return queue    
```

```python
q = Queue()
l = [int(i) for i in input().split(" ")]
for i in range(0,len(l)):
    q.enqueue(l[i])    
ans = Reverse(q)
while ans.isEmpty() is False:
    ele = ans.dequeue()
    print(ele,end=" ")
```
