<h2>Linked list</h2>

```python
class Linked_List:
    def __init__(self,data):
        self.data = data
        self.next = None      
def print_LL(head):
    while head is not None:
        print(str(head.data),"->",end=" ")
        head = head.next
    print("None")     

def input_t():
    input_list = [int(ele) for ele in input().split()]
    head = None
    tail = None
    for curr in input_list:
        if curr == -1:
            break
        newnode = Linked_List(curr)
        if head is None:
            head = newnode
            tail = newnode
        else:
            tail.next = newnode
            tail = newnode
    return head

def length_LL(head):
    count =0
    while head is not None:
        count +=1
        head = head.next
    return count    
head = input_t()
print_LL(head)
length_LL(head)
```

<h2>Print ith Node</h2>

```python
def print_ith(head,i):
    count = 0
    while head is not None:
        if count ==i:
            return head.data
        head = head.next
        count +=1
print_ith(head,1)        
```

<h2>Insert at the Ith Postion</h2>

```python
def insert(head,i,data):
    if i <0 or i>length_LL(head):
        return head
    count = 0
    prev = None
    curr = head
    while count <i:
        prev =curr
        curr = curr.next
        count +=1
    newNode = Linked_List(data)
    if prev is not None:
        prev.next = newNode
    else:
        head = newNode
    newNode.next = curr
    return head
h1 = insert(head,5,9)
print_LL(h1)            

```

<h2>Delete a Node</h2>

```python
def delete(head,pos):
    if pos<0 or pos>length_LL(head):
       return head
    count = 0
    prev = None
    curr = head
    while count <pos:
        prev = curr
        curr = curr.next
        next_ele = curr.next
        count +=1
    if prev is not None:
        prev.next = next_ele
    else:
        head = next_ele
    curr.next = None    
    return head
delete(head,2)
print_LL(head)
```

<h2>Length of LL in Recursive way</h2>

```python
def length_recursive(head):
    if head is None:
        return 0
    return 1 + length_recursive(head.next)   
print(length_recursive(head))
```

<h2>Insert a value at ith position in recursive way</h2>

```python
def insert_recursive(head,i,data):
    if i<0:
        return head
    if i==0:
        newNode = Linked_List(data)
        newNode.next = head
        return newNode   
    if head is None:
        return None
    smallHead = insert_recursive(head.next,i-1,data)
    head.next = smallHead
    return head
```

<h2>Q1 Find a node in linked list</h2>

```python
def findNode(head,n):
    ans = 0
    while head is not None:
        if head.data == n:
            return ans 
        head = head.next
        ans +=1
    return -1    
print(findNode(head,10))         

def print_r(head):
    list = []
    while head is not None:
        list.append(head.data)
        head = head.next
    for i in range(len(list)):
        print(list[-i-1],end=" ")
print_r(head)            
```

<h2>Q3 check Palindrome linked list</h2>

```python
def isPalindrome(head):
    stri = ""
    while head is not None:
        stri += str(head.data)
        head = head.next
    return stri == stri[::-1]         
print(isPalindrome(head))    
```
