plus one

```python
def plusOne(digits):
    s = ""
    for i in digits:
        s +=str(i)
    s1 = int(s)
    s1 +=1
    l = []
    s1 = str(s1)
    for j in s1:
        l.append(int(j))
    return l    
print(plusOne([int(i) for i in input().split()]))
```

pascal triangle

```python
def generate(numRows):
    l = [[1]]
    for i in range(1,numRows+1):
        for j in range(0,i+1):
            if j == 0:
                l.append([1])
            elif i==j:
                l[i].append(1)    
            else:
                l[i].append(l[i-1][j-1] + l[i-1][j])   
    return l[numRows]
n = int(input())    
generate(n)    

```

buy stock and sell

```python
def maxProfit(prices):
    max_prices = 0
    l = []
    short_list = []
    if len(prices) > 1:    
        for i in range(len(prices)):
            for j in range(i,len(prices)-1):
                max_prices = prices[j+1] - prices[i]
                l.append(max_prices)
            if len(l) > 1:
                short_list.append(max(l))
            else:
                short_list.append(l[0])            
        max_prices = max(short_list)        
        if max_prices >= 1:
            return max_prices
        else:
            return 0
    else:
        return 0    
print(maxProfit([7,6,4,3,1]))           
```

majority element

```python
def majorityElement(nums):

    dic = {}
    for i in nums:
        if len(dic) == 0:
            dic[i] = 1
        elif i in dic:
            dic[i] +=1
        else:
            dic[i] = 1   
    return max(dic,key=dic.get) 
majorityElement([2,2,1,1,1,2,2])
```

Merge Sort

```python
def merge_sort(nums1,nums2,m,n):
        for i in nums1:
            if 0 in nums1:
                nums1.remove(0)
        for j in nums2:
            if 0 in nums2:
                nums2.remove(0)    
        m = len(nums1)   
        n = len(nums2)        
        nums3 = []
        i = 0
        j = 0
        while i < m and j < n:
            if nums1[i] < nums2[j]:
                nums3.append(nums1[i])
                i +=1
            elif nums1[i] >=nums2[j]:
                nums3.append(nums2[j])
                j  +=1
        while i < m:
            nums3.append(nums1[i]) 
            i +=1
        while j <n:
            nums3.append(nums2[j])
            j +=1
        nums1 = nums3    
        return nums1                       
arr1 = [0]
m = len(arr1)
arr2 = [1]
n = len(arr2)
merge_sort(arr1,arr2,m,n)             
```

Container with most water

```python
def maxArea(height):
        left = 0 
        right =  len(height)-1
        maxArea = 0
        while left <right:
            currentArea = min(height[left],height[right]) * (right-left)
            maxArea= max(maxArea,currentArea)
            if height[left] < height[right] :
                left +=1
            else:
                right -=1
        return maxArea  
```

3Sum

```python
def threeSum(nums):
        nums.sort()
        dic={}
        lst=[]
        for i in range(len(nums)):
            low=i+1
            high=len(nums)-1
            while low<high:
                sm=nums[i]+nums[low]+nums[high]
                if sm==0:
                    val=[nums[i],nums[low],nums[high]]
                    if str(val) in dic:
                        dic[str(val)]+=1
                    else:
                        dic[str(val)]=1
                        lst.append(val)
                    low+=1
                    high-=1
                elif sm>0:
                    high-=1
                else:
                    low+=1
        return lst
print(threeSum([-1,0,1,2,-1,-4]))
```
