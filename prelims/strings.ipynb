check permutation 

```python
def isPermutation(string1, string2) : 
    if len(string1) != len(string2) : 
        return False 
    frquency = [0] * 256 
    for i in range(len(string1)) : 
        ch = ord(string1[i]) 
        frquency[ch] += 1 
    for i in range(len(string2)) : 
        ch = ord(string2[i]) 
        frquency[ch] -= 1 
    for i in range(256) : 
        if frquency[i] != 0 : 
            return False 
    return True 
string1 = "abcde"
string2 = "baedc"
if isPermutation(string1,string2):
    print("true")
else:
    print("false")    
```

Find the Index of the First Occurrence in a String


```python
def strStr(haystack,needle):
    for i in range(len(haystack)):
        if needle == haystack[i:i+len(needle)] and i < len(haystack):
            return i
    return -1
strStr("hello","ll")    
```

Length of Last Word


```python
def lengthOfLastWord(s):
        li = list(s[::-1])
        count = 0 
        i = 0
        isTrue = True
        while isTrue:
            if li[i] == ' ':
                li.pop(i)
            else:
                isTrue = False 
        for i in li:
            if i == ' ':
                break
            count +=1 
        return count
lengthOfLastWord("   fly me   to   the moon                                         ")
```

valid Palindrome

```python
def isPalindrome(s):
    s1 = ""
    for i in s:
        if i.isalnum():
            s1 +=i
    s2 = s1.lower()
    return s2 == s2[::-1]
isPalindrome("Marge, let's \"[went].\" I await {news} telegram.")                     
```

Longest Palindrome Substring

```python
def longestPalindrome(s):
    s1 = ""
    k = 0
    while k<len(s):
        s2 = s[k:]
        j = len(s2)
        while j>0:
            if s2[k:j]==s2[k:j:-1]:
                s1 = s2[k:j]
                j -=1
        k -=1
    return s1            
```
