Fibonacchi number using recursion

```python
def fibo(n):
    # it takes time complexity O(2^n)
    if n==0 or n==1:
        return n
    return fibo(n-1) + fibo(n-2)
print(fibo(4))    
```

Fibonaachi number using recurison with dp

```python
def fibo_dp(n,dp):
    if n==0 or n==1:
        return n
    if dp[n-1]!=-1:
        ans1 = dp[n-1]
    else:
        dp[n-1] = fibo_dp(n-1,dp)
    if dp[n-2]!=-1:
        ans2 = dp[n-1] 
    ans1 = fibo_dp(n-1,dp)
    ans2 = fibo_dp(n-2,dp)
    return ans1 + ans2
n = int(input())
dp = [-1]*(n+1)
print(fibo_dp(n,dp))
```

Iterative solution of fibinaaci using dp

```python
n = int(input())
dp = [-1]*(n+1)
dp[0] = 0
dp[1] = 1
for i in range(2,n+1):
    dp[i] = dp[i-1] + dp[i-2]
print(dp[n])    
```

minSteps1

```python
def minSteps1(n,ans2=float('inf'),ans3 = float('inf')):
    if n==1:
        return 0
    ans1 = minSteps1(n-1,ans2,ans3)      
    if n%2==0:
        ans2 = minSteps1(n//2,ans2,ans3)
    if n%3==0:
        ans3 = minSteps1(n//3,ans2,ans3)  
    return 1 + min(ans1,ans2,ans3)
n = int(input())    
print(minSteps1(n))
```

minStesps_Recursive using Dp

```python
def minSteps1_dp(n,dp,ans2=float('inf'),ans3 = float('inf')):
    if n==1:
        return 0
    if dp[n-1]==-1:
        ans1 = minSteps1_dp(n-1,dp,ans2,ans3)
        dp[n-1] = ans1
    else:
        ans1 = dp[n-1]  
    if n%2==0:
        if dp[n//2]==-1:
            ans2 = minSteps1_dp(n//2,dp,ans2,ans3)
            dp[n//2]=ans2
        else:
            ans2 = dp[n//2]    
    if n%3==0:
        if dp[n//3]==-1:
            ans3 = minSteps1_dp(n//3,dp,ans2,ans3)
            dp[n//3]=ans3
        else:
            ans3 = dp[n//3]  
    return 1 + min(ans1,ans2,ans3)
n = int(input())  
dp = [-1]*(n)  
print(minSteps1_dp(n,dp))
```

minSteps1_iterative using Dp

```python
def minSteps1_dp_it(n):
    if n==1:
        return 0
    dp = [-1]*(n+1)
    dp[1]=0
    for i in range(2,n+1):
        ans1 = dp[i-1]
        ans2 = float('inf')
        ans3 = float('inf')
        if i%2 ==0:
            ans2 = dp[i//2]
        if i%3==0:
            ans3 = dp[i//3]
        dp[i] = 1 + min(ans1,ans2,ans3) 
    return dp[n] 
n = int(input())
print(minSteps1_dp_it(n))       
```

Perfect Sqaures

```python
def numSquares(n, squares=None, memo=None):
    if not squares:
        squares = [i * i for i in range(1, int(n ** 0.5) + 1)]
        squares.reverse()  # Reverse the squares to start from the largest one
    if not memo:
        memo = {}

    if n == 0:
        return 0

    if n in memo:
        return memo[n]

    min_count = float('inf')
    for square in squares:
        if square > n:
            continue
        count = 1 + numSquares(n - square, squares, memo)
        min_count = min(min_count, count)

    memo[n] = min_count
    return min_count

print(numSquares(13))  # Output: 2
```
