power of three

```python
def isPowerOfThree(n):
    if n == 1:
        return True
    if n <=0 :
        return False    
    if isPowerOfThree(n/3):
        return True
    else:
        return False        
print(isPowerOfThree(3))
```

power of num

```python
def power_num(x,n):
    if n==0:
        return 1
    return x*power_num(x,n-1)
power_num(2,3)
```

check if the list is sorted or not

```python
def is_sorted(l):
    if not l or len(l)==1:
        return True
    small_list = l[1:]
    if l[0]<l[1]:
        is_sorted(small_list)
    return False
is_sorted([1,2,3,4,5])
    
```
