2. Set K Bits (set meaning 0-->1 and for 1 no change)

```python
def setKBits(n, k):
    for i in range(k):
        mask = 1<<i
        n = n|mask
    return n 
setKBits(6,2)
```

Count distinct Bitwise OR of all subarrays

```python
def countDistinctBitwiseOR(arr, size):
    # Write your code here.
    dic = {}
    for i in range(len(arr)):
        bit_or = 0
        for j in range(i,len(arr)):
            bit_or |=arr[j]
            if bit_or in dic:
                dic[bit_or] +=1
            else:
                dic[bit_or]=1

    return len(dic)  
print(countDistinctBitwiseOR([1,2,3],3))
```

check xor of subsets and tip xor of subsets are always 0

```python
def check_xor_subset(nums):
    subsets = [[]]  
    for num in nums:
        current_subsets = []
        for subset in subsets:
            current_subsets.append(subset + [num])
        subsets.extend(current_subsets)
    s1 = subsets[1:]
    xor = 0
    for i in s1:
        sub_xor = 0
        for j in i:
            sub_xor ^=j
        xor ^=sub_xor
    return xor    
nums = [1, 3, 2]
print(check_xor_subset(nums))
```

Generate all subsets using bit manipulation

```python
arr = [int(i) for i in input().split()]
n = len(arr)
l = []
for i in range(0,1<<n):
    l1 = []
    for j in range(0,n):
        if i&(1<<j):
            l1.append(arr[j])
    l.append(l1)    
print(l)
```

Pearl and Combination

```python
def pearl_combination(nums):
    n = len(nums)
    for i in range(0,1<<n):
        sum_ = 0
        for j in range(0,n):
            if i&(1<<j):
                sum_+=nums[j]
            else:
                sum_ -=nums[j] 
        if sum_%360==0:
            return True
    return False               
nums = [int(i) for i in input().split()]
if pearl_combination(nums):
    print('YES')
else:
    print('NO')    
```
