Fibonacci Memonization

```python
# fibonacci normal recursive solution
def fibo(n):
    if n==0 or n==1:
        return n
    return fibo(n-1) + fibo(n-2) 
fibo(int(input()))    
```

```python
# Fibonacci minSteps recursive solution
def fibo_minSteps(n,minSteps):
    if n==0 or n==1:
        return n
    if minSteps[n-1] ==-1:
        ans1 = fibo_minSteps(n-1,minSteps)
        minSteps[n-1] = ans1
    else:
        ans1 = minSteps[n-1]
    if minSteps[n-2]==-1:
        ans2 = fibo_minSteps(n-2,minSteps)
        minSteps[n-2] = ans2
    else:
        ans2 = minSteps[n-2]
    return ans1 + ans2                
n = int(input())
minSteps = [-1]*n
fibo_minSteps(n,minSteps)
```

N-th Tribonacci Number

```python
def tribonacci(n):
    if n==0 or n==1:
        return n
    if n==2:
        return 1
    return tribonacci(n-1) + tribonacci(n-2) + tribonacci(n-3)
tribonacci(31)
```

minSteps SOLUTION FOR Tribonacci 

```python
def tribonaaci_minSteps(n,minSteps):
    if n==0 or n==1:
        return n
    if n==2:
        return 1
    if minSteps[n-1]==-1:
        ans1 = tribonaaci_minSteps(n-1,minSteps)
        minSteps[n-1] = ans1
    else:
        ans1 = minSteps[n-1]    
    if minSteps[n-2]==-1:
        ans2 = tribonaaci_minSteps(n-2,minSteps)
        minSteps[n-2] = ans2
    else:
        ans2 = minSteps[n-2]    
    if minSteps[n-3]==-1:
        ans3 = tribonaaci_minSteps(n-3,minSteps)
        minSteps[n-3] = ans3
    else:
        ans3 = minSteps[n-3]           
    return ans1 + ans2 + ans3
n = int(input())
minSteps = [-1]*n     
tribonaaci_minSteps(n,minSteps)
```

Minimum Steps To 1

```python
import sys
def minStep(n):
    if n==1:
        return 0
    ans1 = minStep(n-1)
    ans2 = sys.maxsize
    ans3 = sys.maxsize
    if n%2==0:
        ans2 = minStep(n//2)
    if n%3 ==0:
        ans3 = minStep(n//3)    
    ans = 1 + min(ans1,ans2,ans3)  
    return ans  
minStep(4)
```

```python
# minSteps SOLUTIONS FOR THE SAME
from sys import maxsize as MAX_VALUE
def minStep_minSteps(n,minSteps):
    if n==1:
        return 0
    ans2 = MAX_VALUE
    ans3 = MAX_VALUE
    if minSteps[n-1]==-1:
        ans1 = minStep_minSteps(n-1,minSteps)
        minSteps[n-1] = ans1
    else:
        ans1 = minSteps[n-1]    
    if n%2==0:
        if minSteps[n//2] ==-1:   
            ans2 = minStep_minSteps(n//2,minSteps)
            minSteps[n//2] = ans2
        else:
            ans2 = minSteps[n//2]    
    if n%3 ==0:
        if minSteps[n//3]==-1:
            ans3 = minStep_minSteps(n//3,minSteps)  
            minSteps[n//3]=ans3  
        else:
            ans3 = minSteps[n//3]    
    ans = 1 + min(ans1,ans2,ans3)  
    return ans
n = int(input())
minSteps = [-1]*(n+1)
print(minStep_minSteps(n,minSteps))
```

```python
# minSteps Iterative solution
import sys
def minStep_Iter_minSteps(n):
    if n==1:
        return 0
    minSteps= [0]*(n+1)
    minSteps[1] = 0
    for currStep in range(2,n+1):
        subractone = minSteps[n-1]
        divideByTwo =sys.maxsize
        divideByThree = sys.maxsize
        if currStep %2==0:
           divideByTwo = minSteps[currStep//2]
        elif currStep%3==0:
            divideByThree = minSteps[currStep//3]
        minSteps[currStep] = 1 + min(subractone,divideByTwo,divideByThree) 
    return minSteps[n] 
n =int(input())          
print(minStep_Iter_minSteps(n+1)) 
  


```

Interger Break

```python
def integerBreak(n):
    if n==0:
        return 0
    if n <= 2:
        return 1
    if n == 3:
        return 2
    dp = [0] * (n + 1)
    dp[2] = 2
    dp[3] = 3
    for i in range(4, n + 1):
        dp[i] = max(2 * dp[i - 2], 3 * dp[i - 3])
    return dp[n]
n = int(input())
integerBreak(n)
```

Min Cost Climbing Stairs

```python
def minCostClimbingStairs(cost) -> int:
    dp = [0] * len(cost)
    dp[0] = cost[0]
    dp[1] = cost[1]
    for i in range(2, len(cost)):
        dp[i] = cost[i] + min(dp[i - 1], dp[i - 2])
    return min(dp[-1], dp[-2])
cost = [1,100,1,1,1,100,1,1,100,1]    
minCostClimbingStairs(cost) 
```
