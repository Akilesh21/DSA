Calculate the minimum sum of K,consecutive elements

```python
def min_sum_k(arr,k):
    sum_k = sum(arr[0:k])
    for i in range(1,len(arr)):
        sum_k =  min(sum_k,sum_k-arr[i-1]+arr[i+k-1])  
    return sum_k         
min_sum_k([int(i) for i in input().split()],int(input()))
```
