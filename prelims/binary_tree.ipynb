Second Minimum Node in a Binary Tree

```python
def array_bt(root,array=[]):
    if root is None:
        return 
    array.append(root.data)
    array_bt(root.left)
    array_bt(root.right)
    return array
def findSecondMinimumValue(root):
    arr = array_bt(root)
    max_ele = max(arr)
    if max_ele in arr:
        arr.remove(max_ele)
    if arr.count(max_ele) == len(arr):
        return -1
    else:
        return max(arr)    
b6 = treeinput()
print(findSecondMinimumValue(b6))
```

Balanced Tree

```python
def Height(root):
    if root is None:
        return 0
    return 1 + max(Height(root.left),Height(root.right))    
def isBalanced(root):
    if root is None:
        return True
    lh = Height(root.left)
    rh = Height(root.right)
    if lh - rh >1 or rh - lh>1:
        return False
    if isBalanced(root.left) and isBalanced(root.right):
        return True
    else:
        return False
b10 = treeinput()
isBalanced(b10)
```

Diameter of Binary Tree

```python
def height(root):
    if root is None:
        return 0
    lh = height(root.left)
    rh = height(root.right)
    return 1 + max(lh,rh)
def diameterOfBinaryTree(root) :
    if root ==None:
        return 0
    if root.left!=None:
        return 1+height(root.left) + height(root.right)
    elif root.left ==None:
        rd = diameterOfBinaryTree(root.right)
        return 1+rd
    elif root.right ==None:
        ld=diameterOfBinaryTree(root.left)
        return 1+ld    
b11 = treeinput()
diameterOfBinaryTree(b11)
```

Levelwise Input

```python
import queue
def levelwise_input():
    q = queue.Queue()
    data = int(input())
    if data ==-1:
        return None
    root = BinaryTreeNode(data)
    q.put(root)
    while(not(q.empty())):
        current_data = q.get()
        leftchild_data = int(input())
        if leftchild_data !=-1:
            leftchild = BinaryTreeNode(leftchild_data)
            current_data.left = leftchild
            q.put(leftchild)
        rightchild_data = int(input())  
        if rightchild_data !=-1:
            rightchild = BinaryTreeNode(rightchild_data)
            current_data.right = rightchild
            q.put(rightchild)  
    return root        
```

```python
from os import *
from sys import *
from collections import *
from math import *
import queue
# BinaryTreeNode class definition
# class BinaryTreeNode:
#     def __init__(self, data):
#         self.data = data
#         self.left = None
#         self.right = None      
def zigZagTraversal(root,level=0):
    if root is None:
        return 
    q = queue.Queue()
    q.put(root)
    l = []
    while (not(q.empty())):
        level_size = q.qsize()
        if level%2==0:
            l1 = []
            for i in range(q.qsize()):
                curr = q.get()
                l1.append(curr.data)
                if curr.left is not None:
                    q.put(curr.left)
                if curr.right is not None:
                    q.put(curr.right)  
            l.append(l1)                 
        else:
            l2 = []
            for i in range(q.qsize()):
                curr = q.get()
                l2.append(curr.data)
                if curr.left is not None:
                    q.put(curr.left)
                if curr.right is not None:
                    q.put(curr.right)    
            l.append(l2[::-1])                
        level +=1        
    return l


root = levelwise_input()
zigZagTraversal(root)                   
```

Construct binary Tree using Inorder and postorder

```python
def buildTree1(inorder, postorder):
    if not inorder or not postorder:
        return None

    root_val = postorder.pop()
    root = BinaryTreeNode(root_val)
    inorder_index = inorder.index(root_val)

    # Build right subtree first, as elements are popped from the end
    root.right = buildTree1(inorder[inorder_index + 1:], postorder)
    root.left = buildTree1(inorder[:inorder_index], postorder)

    return root
inorder = [9, 3, 15, 20, 7]
postorder = [9, 15, 7, 20, 3]
root = buildTree1(inorder, postorder)
printLevelWise(root)
```

Kth Smallest Element in a BST

```python
def kthSmallest(root, k) :
    q = queue.Queue()
    q.put(root)
    l = []
    while (not(q.empty())):
        level_size = q.qsize()
        for i in range(level_size):
            curr = q.get()
            if curr.data==k:
                l.append(curr.data)
                break
            else:
                l.append(curr.data) 
            if curr.left is not None:
                q.put(curr.left)
            if curr.right is not None:
                q.put(curr.right)
    return min(l)   
root = levelwise_input()
print(kthSmallest(root,2))            
```
