Binary Tree Practice


```python
class BinaryTreeNode:
    def __init__(self,data):
        self.data = data
        self.left = None
        self.right = None
def print_BT(root):
    if root ==None:
        return 
    print(root.data,end=":")
    if root.left !=None:
        print("L",root.left.data,end=",")
    if root.right !=None:
        print("R",root.right.data) 
    print()    
    print_BT(root.left)
    print_BT(root.right)         
def BT_input():
    rootData = int(input())
    if rootData == -1:
        return None
    root = BinaryTreeNode(rootData)
    rootleft = BT_input()
    rootright = BT_input()
    root.left = rootleft
    root.right = rootright
    return root
            
```

```python
root = BT_input()
print_BT(root)
```

```python
def num_of_nodes(root):
    if root is None:
        return 0
    leftcount = num_of_nodes(root.left)  
    rightcount = num_of_nodes(root.right)
    return 1 + leftcount + rightcount
num_of_nodes(root)    
```

```python
def sum_of_nodes(root):
    if root is None:
        return 0
    leftdata = sum_of_nodes(root.left)
    rightdata = sum_of_nodes(root.right)
    return root.data + leftdata + rightdata
sum_of_nodes(root)
```

```python
def nodes_greater_x(root,x):
    if root ==None:
        return 0
    if root.data < x :
        value =  1 + nodes_greater_x(root.left,x) + nodes_greater_x(root.right,x)
    elif root.data > x :
        nodes_greater_x(root.left,x)
        nodes_greater_x(root.right,x)
    return value
print(nodes_greater_x(root,2))
```
